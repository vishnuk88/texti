```
import pandas as pd 
from sklearn.feature_extraction.text import TfidfTransformer,CountVectorizer
from nltk.stem import WordNetLemmatizer 
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import cdist
import seaborn as sns
import nltk
nltk.download('wordnet') 
```

```
df = pd.read_csv("/content/drive/MyDrive/runningMate/export_dataframe.csv")
df.head()
```

```
df = df.replace('₹','', regex=True).replace(',','',regex=True)
df['price'] =df['price'].astype(str).astype(float)
df.dtypes
```

```
 nltk.download('stopwords')
stop = stopwords.words('english')
df['NoStopWordName'] = df['name'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
df['NoStopWordName'] = df['NoStopWordName'].astype(str)
df.head()
```

```
df['_any'] = df['name'].str.len() - df['NoStopWordName'].str.len()
df.query('_any > 0')
```
